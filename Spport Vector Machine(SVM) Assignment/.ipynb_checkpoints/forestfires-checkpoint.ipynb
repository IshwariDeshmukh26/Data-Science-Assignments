In California, annual forest fires can cause huge loss of wildlife, human life, and can cost billions of dollars in property damage. Local officials would like to predict the size of the burnt area in forest fires annually so that they can be better prepared in future calamities. 
Build a Support Vector Machines algorithm on the dataset and share your insights on it in the documentation. 
Note: - Size_ category is the output variable.


## business objective

To predict the size category of forest fires (small or large) based on weather and environmental factors to help authorities take proactive measures in fire prevention and management.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

## Step 2: Load Dataset

```python
df = pd.read_csv("forestfires.csv") 
df
```

## Step 3: Data Pre-processing

### 3.1 Data Cleaning & Feature Engineering

```python
# Drop irrelevant or unused columns
df.drop(['area'], axis=1, inplace=True)
```

```python
# Encode target variable
df['size_category'] = df['size_category'].map({'small': 0, 'large': 1})
```

```python
# One-hot encode categorical features
df = pd.get_dummies(df, columns=['month', 'day'], drop_first=True)
df
```

### 3.2 Outlier Treatment (IQR Method)

```python
def cap_outliers(col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR
    df[col] = np.where(df[col] < lower, lower, df[col])
    df[col] = np.where(df[col] > upper, upper, df[col])

for col in ['FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind', 'rain']:
    cap_outliers(col)
```

##  Step 4: EDA (Exploratory Data Analysis)

### 4.1 Summary

```python
df.describe()
```

### 4.2 Univariate Analysis

```python
for col in ['FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind', 'rain']:
    sns.histplot(df[col], kde=True)
    plt.title(f'{col} Distribution')
    plt.show()
```

### 4.3 Bivariate Analysis

```python
plt.figure(figsize=(14, 10))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

### Step 5: Model Building (SVM)

### 5.1 Scaling

```python
X = df.drop('size_category', axis=1)
y = df['size_category']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

### 5.2 Train-Test Split



```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```

### 5.3 SVM Model Training and Evaluation

```python
kernels = ['linear', 'rbf', 'poly', 'sigmoid']

for kernel in kernels:
    model = SVC(kernel=kernel)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)

    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
    print("Classification Report:\n", classification_report(y_test, y_pred))
    print("\n")
```

## Step 6: Model Explanation

The SVM algorithm was tested with various kernels to find the best model for classification. 
The confusion matrix and accuracy score provide insights into model performance.
